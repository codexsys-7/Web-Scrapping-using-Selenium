```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AVN.csv')
df.head()
```

```python
df.shape
```

```python
df[df['Volume'] > 300000]
```

```python
df.isnull().sum()
```

```python
#ax= plt.figure(figsize =(12,8))
```

```python
plt.figure(figsize = (12,7))
df['Time'] = pd.to_datetime(df.Time)
sns.lineplot(df['Time'], df['Close'], legend = 'full')
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.plot(df['Time'], df['Close'])
plt.legend('Close')
plt.show()
```

```python
days = 25
df['new_col'] = df['Close'].rolling(days).mean()
df.head()
```

```python
plt.figure(figsize =(12,7))
sns.lineplot(df['Time'], df['new_col'])
sns.lineplot(df['Time'], df['Close'])
plt.legend('Close')
plt.show()
```

```python
days = [10, 20, 30]
for i in days:
    col_name = 'my_avg_'+str(i)+'_days'
    df[col_name] = df['Close'].rolling(i).mean()
```

```python
df.head(20)
```

```python
plt.figure(figsize = (18,12))
sns.lineplot(df['Time'], df['Close'])
sns.lineplot(df['Time'], df['new_col'])
sns.lineplot(df['Time'], df['my_avg_10_days'])
sns.lineplot(df['Time'], df['my_avg_20_days'])
sns.lineplot(df['Time'], df['my_avg_30_days'])
plt.plot()
```

# Daily Returns

```python
df['daily_open'] = df['Open'].pct_change()
df['daily_close'] = df['Close'].pct_change()
```

```python
plt.figure(figsize =(12,8))
sns.lineplot(df['Time'], df['daily_open'])
plt.show()
```

```python
plt.figure(figsize =(12,8))
sns.lineplot(df['Time'], df['daily_close'])
plt.show()
```

```python
plt.figure(figsize =(12,8))
sns.distplot(df['daily_open'].dropna(), color ='blue')
plt.show()
```

```python
plt.figure(figsize =(12,8))
sns.distplot(df['daily_close'].dropna(), color ='blue')
plt.show()
```
